```python
# .loc[]
# Update single value

import pandas as pd

data = {
    "Name" : ['Rahul','Debu', 'Dev', 'Subho', 'Anshu', 'Simran', 'Souvik', 'Satak'],
    "Age" : [20,20,22,24,24,20,22,23],
    "Salary" : [70000, 40000, 50000, 45000, 35000, 30000, 80000, 60000],
    "Performance Score" : [95, 75, 80, 78, 72, 70, 89, 88]
} 

df = pd.DataFrame(data)

df.loc[0,"Salary"] = 85000
print(df)
```

```python
# updating multiple values

import pandas as pd

data = {
    "Name" : ['Rahul','Debu', 'Dev', 'Subho', 'Anshu', 'Simran', 'Souvik', 'Satak'],
    "Age" : [20,20,22,24,24,20,22,23],
    "Salary" : [70000, 40000, 50000, 45000, 35000, 30000, 80000, 60000],
    "Performance Score" : [95, 75, 80, 78, 72, 70, 89, 88]
} 

df = pd.DataFrame(data)

df['Salary'] = df['Salary']* 1.25
print(df)
```
