```python
'''groupby() is a Pandas method used to group data based on one or more columns and perform 
    operations such as sum, mean, count, max, min, etc. on each group.'''

import pandas as pd

data = {
    "Department": ["IT", "HR", "IT", "HR", "Sales", "IT"],
    "Employee": ["Rahul", "Amit", "Dev", "Anshu", "Souvik", "Riya"],
    "Salary": [50000, 40000, 60000, 45000, 35000, 70000]
}

df = pd.DataFrame(data)
print(df.groupby("Department")['Salary'].mean())
```
