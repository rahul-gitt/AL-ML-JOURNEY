```python
# Inner join

import pandas as pd

students = pd.DataFrame({
    'ID' : [100,101,103],
    'Name': ['Rahul', 'Dev','Debu']
})

marks = pd.DataFrame({
    'ID': [100,101,104],
    'Marks': [90,80,86]
})

result = pd.merge(students,marks,on="ID",how= "inner")
print(result)
```

```python
# outer join

import pandas as pd

students = pd.DataFrame({
    'ID' : [100,101,103],
    'Name': ['Rahul', 'Dev','Debu']
})

marks = pd.DataFrame({
    'ID': [100,101,104],
    'Marks': [90,80,86]
})

result = pd.merge(students,marks,on="ID", how= "outer")
print(result)
```

```python
#left join

import pandas as pd

students = pd.DataFrame({
    'ID' : [100,101,103],
    'Name': ['Rahul', 'Dev','Debu']
})

marks = pd.DataFrame({
    'ID': [100,101,104],
    'Marks': [90,80,86]
})

result = pd.merge(students,marks, on="ID", how="left")
print(result)
```

```python
# right join 
import pandas as pd

students = pd.DataFrame({
    'ID' : [100,101,103],
    'Name': ['Rahul', 'Dev','Debu']
})

marks = pd.DataFrame({
    'ID': [100,101,104],
    'Marks': [90,80,86]
})

result= pd.merge(students, marks, on= "ID", how= "right")
print(result)
```
