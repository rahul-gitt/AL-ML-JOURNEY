```python
# concatination()
# vertical(row wise)

import pandas as pd
df1 = pd.DataFrame({
    'Name' : ['Rahul','Dev'],
    'Age' : [19,22]
})

df2 = pd.DataFrame({
    'Name': ['Debu','Subho'],
    'Age': [20,24]
})

result = pd.concat([df1,df2],ignore_index=True)
print(result)
```

```python
# horizontal (column wise)

import pandas as pd
df1 = pd.DataFrame({
    'Name' : ['Rahul','Dev'],
    'Age' : [19,22]
})

df2 = pd.DataFrame({
    'Name': ['Debu','Subho'],
    'Age': [20,24]
})
result = pd.concat([df1,df2],axis=1, ignore_index=True)
print(result)
```
