```python
# Merging
'''merge() is a Pandas method used to combine two DataFrames based on one or more common columns.'''

import pandas as pd

students = pd.DataFrame({
    'ID' : [100,101,103],
    'Name': ['Rahul', 'Dev','Debu']
})

marks = pd.DataFrame({
    'ID': [100,101,103],
    'Marks': [90,80,86]
})

result = pd.merge(students,marks,on="ID")
print(result)
```
