```python
# method - 1
# Add column via assignment

import pandas as pd
data = {
    "Name" : ['Rahul','Debu', 'Dev', 'Subho', 'Anshu', 'Simran', 'Souvik', 'Satak'],
    "Age" : [20,20,22,24,24,20,22,23],
    "Salary" : [70000, 40000, 50000, 45000, 35000, 30000, 80000, 60000],
    "Performance Score" : [95, 75, 80, 78, 72, 70, 89, 88]
} 

df = pd.DataFrame(data)

print("Before adding any column : ")
print(df)

df["Bonus"] = df["Salary"]*0.1
print("After adding column : ")
print(df)
```

```python
# method - 2
# Add column via insert method

import pandas as pd

data = {
    "Name" : ['Rahul','Debu', 'Dev', 'Subho', 'Anshu', 'Simran', 'Souvik', 'Satak'],
    "Age" : [20,20,22,24,24,20,22,23],
    "Salary" : [70000, 40000, 50000, 45000, 35000, 30000, 80000, 60000],
    "Performance Score" : [95, 75, 80, 78, 72, 70, 89, 88]
} 

df = pd.DataFrame(data)

df["Bonus"] = df["Salary"]*0.1
print(df)

print("After adding column : ")
df.insert(2, "ID", [100,101,102,103,104,105,106,107])
print(df)
```
