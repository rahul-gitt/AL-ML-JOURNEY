```python
# fillna()

import pandas as pd
data = {
        "Name" : ['Rahul',"Debu", 'Dev', 'Subho', 'Anshu', 'Simran', 'Souvik', 'Satak'],
        "Age" : [20,20,None,24,24,20,22,23],
        "Salary" : [70000, 40000, 50000, 45000, None, 30000, 80000, 60000],
        "Performance Score" : [95, 75, 80, 78, None, 70, 89, 88]
    } 

df = pd.DataFrame(data)

df.fillna(0, inplace= True)
print(df)
```

```python
# fill data via mean method

import pandas as pd
data = {
        "Name" : ['Rahul',"Debu", 'Dev', 'Subho', 'Anshu', 'Simran', 'Souvik', 'Satak'],
        "Age" : [20,20,None,24,24,20,22,23],
        "Salary" : [70000, 40000, 50000, 45000, None, 30000, 80000, 60000],
        "Performance Score" : [95, 75, 80, 78, None, 70, 89, 88]
    } 

df = pd.DataFrame(data)

df["Age"] = df["Age"].fillna(df["Age"].mean())

df['Performance Score'] = df['Performance Score'].fillna(df['Performance Score'].mean())

df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
print(df)
```
