```python
'''
interpolate()
interpolate() is a Pandas method used to fill missing values (NaN)
 by estimating them from the neighboring data instead of using a fixed value.'''

import pandas as pd

data = {
        "Name" : ['Rahul',"Debu", 'Dev', 'Subho', 'Anshu', 'Simran', 'Souvik', 'Satak'],
        "Age" : [20,20,None,24,24,20,22,23],
        "Salary" : [70000, 40000, 50000, 45000, None, 30000, 80000, 60000],
        "Performance Score" : [95, 75, 80, 78, None, 70, 89, 88]
    } 

df = pd.DataFrame(data)

df['Salary'] = df['Salary'].interpolate()

df['Performance Score'] = df['Performance Score'].interpolate()

df['Age'] = df['Age'].interpolate()
print(df)
```
