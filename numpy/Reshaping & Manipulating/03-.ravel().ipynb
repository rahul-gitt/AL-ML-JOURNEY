```python
'''Ravel in NumPy:

ravel() is used to convert a multi-dimensional array 
into a one-dimensional (1D) array.

It returns a view of the original array whenever possible, 
instead of creating a new copy.'''

import numpy as np 
arr = np.array([
    [10,20,30],
    [40,50,60]
])

new_arr = arr.ravel()

new_arr[0] = 100
print(new_arr)
```
