```python
'''Reshaping in NumPy:

Reshaping means changing the shape (dimensions) of an array 
without changing its original data.

It allows us to convert arrays from one shape to another 
such as 1D → 2D or 2D → 3D.

The total number of elements must remain the same before 
and after reshaping.'''

import numpy as np
arr = np.array([10,20,30,40,50,60])

new_arr = arr.reshape(2,3)
print(new_arr)
```
