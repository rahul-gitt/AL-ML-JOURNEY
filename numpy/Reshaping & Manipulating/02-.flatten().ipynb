```python
'''Flattening in NumPy:

Flattening means converting a multi-dimensional array 
(2D, 3D, etc.) into a one-dimensional (1D) array.

It keeps all the original elements but changes the 
structure into a single row.'''

import numpy as np
arr = np.array([
    [1,2,3],
    [4,5,6]
])

new_arr = arr.flatten()
print(new_arr)
```
