```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("netflix_titles (1).csv")

df['date_added']= pd.to_datetime(df['date_added'],errors= "coerce")

type_count= df['type'].value_counts()
plt.bar(type_count.index,type_count.values,color= ['blue', 'green'])
plt.xlabel("Types of shows")
plt.ylabel("Number of shows")
plt.title('Movies and TV shows chart')
plt.show()

year_count= df.groupby("release_year").size()
plt.plot(year_count.index,
          year_count.values,
         label= "Movies per year", 
         color = 'red', 
         linestyle= '--'
)

plt.grid(True, linestyle= ":", alpha= 0.5, linewidth= 1)

plt.title("Year Wise Movie Realise Chart")
plt.xlabel("Release Year")
plt.ylabel("Number of Shows")
plt.legend(loc = 'best', fontsize= 10)
plt.tight_layout()
plt.show()

country_count = df.groupby("country").size().nlargest(10)
plt.barh(country_count.index, 
         country_count.values,
         color= 'orange',
         edgecolor= 'black')
plt.title("Top 10 Countries with most Netflix Content")
plt.xlabel("Number of Content")
plt.ylabel("Name of the Country")

plt.tight_layout()
plt.show()

calculate_per = df['type'].value_counts()
plt.pie(
    calculate_per.index, labels= calculate_per.values,
    colors= ['red', 'yellow'],
    autopct= '%1.1f%%',
    shadow= True
)

plt.title("Movies & TV Shows pie chart")
plt.show()
```
