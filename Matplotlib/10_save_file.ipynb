```python
import matplotlib.pyplot as plt 

x = ['Rahul', 'Debu', 'Dev', 'Subho', 'Anshu']
y = [80, 65, 76, 95, 69]

plt.plot(x, y, color = 'blue', label= "Student's marks", linestyle= '--', marker= 'o')
plt.xlabel("<- Names ->")
plt.ylabel('<- marks ->')
plt.title("2026 studens marks graph")
plt.legend(loc ='best', fontsize =9)
plt.grid(color = 'gray', linewidth= 1, linestyle = ":")

plt.savefig("simple_graph.png", dpi= 250, bbox_inches= "tight")
plt.show()
```
