```python
# Single Scatter ploting.

import matplotlib.pyplot as plt

study_hr = [1,2,3,4,5,6,7,8]
marks = [45,50,55,60,65,70,75,90]

plt.scatter(study_hr, marks, color= 'blue', marker= 's', label = 'Marks Chart')

plt.title("Students Study hr. and Marks Table")
plt.xlabel("Study hours")
plt.ylabel("Marks Obtain")
plt.legend(loc= 'best', fontsize= 9)
plt.grid(True, linestyle= ':', linewidth = 1, color= 'red')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter([1,2,3,4], [50,60,70, 80], label= 'Class A', color= 'orange')
plt.scatter([1,2,3,4], [45,55,65,75], label = 'Class B', color= 'red', marker= '^')

plt.title('Comparison of two Class')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.legend()
plt.grid(True)

plt.show()
```
