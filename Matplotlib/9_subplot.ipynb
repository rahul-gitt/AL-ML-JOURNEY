```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [10,20,15,25]
 
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("Line chart")

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title("Bar chart")
plt.show()
```

```python
# advanced subplot 
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1,2,figsize = (11,5))

x = [1,2,3,4]
y= [10,20,15,25]

ax[0].plot(x,y, color = 'blue')
ax[0].set_title('Line chart')

ax[1].bar(x,y, color= 'green')
ax[1].set_title("Bar chart")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1,2, figsize= (10,5))

explode = [0.1, 0,0 ,0]

x = [1,2,3,4]
y = [15,9,25, 35]

ax[0].pie(x, labels =y, 
          colors= ['red', 'blue', 'green', 'coral'],
          shadow = True,
          explode = explode)
ax[0].set_title('Pie chart')

ax[1].plot(x,y, color = 'red')
ax[1].set_title('Line chart')
fig.suptitle("Comparision of Line and  pie chart")

plt.tight_layout()
plt.show()
```
