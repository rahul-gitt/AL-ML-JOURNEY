```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apl", "May"]
sales = [15000,27000,5000,19000,13000]

plt.plot(months, sales, color= "red", linestyle="-.", marker= "*", label= "2026 Sales Data")

plt.xlabel("<-Months->")
plt.ylabel("<-Sales->")
plt.title('Monthly Sales Data Repot ')
plt.legend(loc = 'lower right', fontsize= 10)

plt.show()
```
