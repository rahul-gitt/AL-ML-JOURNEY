```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apl", "May"]
sales = [15000,25000,5000,19000,13000]

plt.plot(months, sales, color= "red", linestyle="-.", marker= "*", label= "2026 Sales Data")

plt.xlabel("<-Months->")
plt.ylabel("<-Sales->")
plt.title('Monthly Sales Data Repot ')
plt.legend(loc = 'best', fontsize= 10)
plt.grid(color= "gray", linestyle= ":", linewidth= 1)

plt.show()
```

```python
import  matplotlib.pyplot as plt

month = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [500, 1500,1000,700,2000]

plt.plot(month, sales, label = "2026 Sales Data", linestyle= "--", marker= "o", color= "blue")

plt.title("Monthly Sales Report")
plt.xlabel("<-  Months  ->")
plt.ylabel("<-  Sales  ->")

plt.legend(loc= "best", fontsize= 9)
plt.grid(color= "red", linestyle= ":", linewidth= 1)

plt.xlim(1,4)
plt.ylim(0,2000)

plt.show()
```
