```python
import matplotlib.pyplot as plt

regions= ['North', 'South', 'East', 'West']
sales= [20000, 7000, 26900, 9580]

plt.pie(sales, labels = regions, colors= ['gold', 'skyblue', 'lightgreen', 'coral'], autopct= '%1.1f%%')
plt.title("Region wise sales")
plt.show()
```

```python
import matplotlib.pyplot as plt

departments = ["CSE", "AIML", "ECE", "BBA"]
students = [40, 30, 20, 10]

explode = [0.1, 0, 0, 0]

plt.pie(
    students,
    labels=departments,
    autopct="%1.1f%%",
    colors=["red", "blue", "green", "orange"],
    explode=explode,
    shadow=True,
    startangle=90
)

plt.title("Students Distribution")

plt.show()
```
