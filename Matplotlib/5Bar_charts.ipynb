```python
import matplotlib.pyplot as plt

students = ['Rahul', 'Anshu', 'Dev', 'Debu', 'Subho']
marks= [84, 75, 95, 70, 79]

plt.bar(students, marks, label= "Marks chart", color= "orange", )
plt.title("2026 Students marks chart")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.legend(loc= 'best', fontsize= 9)

plt.show()
```
