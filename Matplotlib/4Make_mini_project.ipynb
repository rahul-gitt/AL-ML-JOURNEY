```python
import matplotlib.pyplot as plt

persons = int(input("Enter how many people you want : "))

names = []
marks = []

for i in range (persons):
    names.append(input("Enter name : "))
    marks.append(int(input("Enter marks : ")))

plt.plot(names, marks, color= "blue", linestyle= "--", marker= "o", label ="3rd sem marks")

plt.title("Marks Graph")
plt.xlabel("<- Names ->")
plt.ylabel("<- Marks ->")

plt.legend(loc= 'best', fontsize= 8)
plt.grid(color= 'gray', linewidth=1, linestyle= ":")

plt.show()
```
